```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv('C:/Users/DEBRA/Desktop/My DataSets/sample.csv')
```

```python
df.head(50)
```

```python
imputer = SimpleImputer(strategy = 'constant')
df2 = pd.DataFrame(imputer.fit_transform(df), columns=df.columns)
```

```python
df2
```

```python
Coffees = df2['Coffee'].unique()
len(Coffees)
```

```python
for idx in range(0,len(Coffees)):
    C = df2[df2['Coffee']==Coffees[idx]].reset_index()
    plt.scatter(np.arange(0,len(C)),C['Year'], color ='b', label ='Year')
    plt.scatter(np.arange(0,len(C)),C['Barley'], color ='r', label ='Barley')
    plt.scatter(np.arange(0,len(C)),C['Cotton'], color ='k', label ='Cotton')
    plt.title(Coffees[idx])
    plt.xlabel('Days...')
    plt.ylabel('hhhhh')
    plt.legend()
    plt.show()
    
```
